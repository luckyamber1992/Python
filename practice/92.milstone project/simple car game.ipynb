```python
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
# create a card class to assosicate suits,ranks and values
class card:
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
        self.value=values[self.rank]

    def __str__(self):
        return f'{self.rank} of {self.suit}'
                
```

```python
# create a sample card
my_card=card(suits[1],ranks[1])
print(my_card)
my_card.value
```

```python
# create a deck. In which we will use card class
class Deck:
    def __init__(self):
        self.all_cards=[]
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(card(suit,rank))
    def shuffle(self):
        random.shuffle(self.all_cards)
    
    def del_one(self):
        return self.all_cards.pop()

    def __str__(self):
        return f'there is {len(self.all_cards)} in the deck'
```

```python
# create player class. a player should be able to remove from top of the list and add cards 
class player:
    
    def __init__(self,name):
        self.name=name
        self.all_cards=[]

    def add_cards(self,new_cards):
        if type(new_cards)==type([]):
            self.all_cards.extend(new_cards)
        else:
            self.all_cards.append(new_cards)
    def remove_one(self):
        return self.all_cards.pop(0)
    
    def __str__(self):
        return f'{self.name} has {len(self.all_cards)} cards'
```
